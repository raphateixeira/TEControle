```python
import numpy as np
import plotly.express as px
```

O sistema de Aizawa

```python
a = 0.95
b = 0.7
c = 0.6
d = 3.5
e = 0.25
f = 0.1
```

$$ 
\begin{align*}
    \frac{dx}{dt} = &(z-b)x - dy\\
    \frac{dy}{dt} = &dx + (z-b)y\\
    \frac{dz}{dt} = &c + az - \frac{z^3}{3} -\\ &(x^2+y^2)(1+ez) + fzx^3
\end{align*}
$$

```python
def aizawa(x0, t):
    x, y, z = x0
    dx = (z-b)*x-d*y
    dy = d*x + (z-b)*y
    dz = c + a*z - z**3/3 - (x**2+y**2)*(1+e*z)+f*z*x**3

    return np.array([dx,dy,dz])
```

```python
from scipy.integrate import odeint
```

```python
t = np.linspace(0.0,100.0,10000)
x0 = np.array([0.1, -0.5, 0.6])
x = odeint(aizawa, x0, t)
```

```python
fig = px.line_3d(x = x[:,0],y = x[:,1],z = x[:,2])
fig.update_layout(width=1000, height=800)
fig.show()
```

```python
fig1 = px.line(x=t, y = x[:,0])
fig1.show()
```
